```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
df = pd.read_csv('adult.csv') # Ensure the filename matches yours

# Quick look at categorical and numerical features (Hint 1)
print(df.info())
```

```python
# 2. Apply Label Encoding for 'education' (example of ordered data)
le = LabelEncoder()
df['education'] = le.fit_transform(df['education'])

# 3. Apply One-Hot Encoding for 'marital-status' and 'gender'
df = pd.get_dummies(df, columns=['marital-status', 'gender'], drop_first=True)
```

```python
# 4. Scale numerical features
scaler = StandardScaler()

# Compare readiness before scaling
print("Before Scaling:\n", df[['age', 'hours-per-week']].head())

# Apply Scaling
df[['age', 'hours-per-week']] = scaler.fit_transform(df[['age', 'hours-per-week']])

# 5. Compare after scaling
print("\nAfter Scaling:\n", df[['age', 'hours-per-week']].head())
```

```python
# 7. Save processed dataset
df.to_csv('Processed_Adult_Income.csv', index=False)
```
